个变量。唯一的前提就是变量的数量必须跟序列元素的数量是一样的。

包括字符串,文件对象,迭代器和生成器。
 - 有时候,你可能只想解压一部分,丢弃其他的值。对于这种情况 Python 并没有提
供特殊的语法。但是你可以使用任意变量名去占位,到时候丢掉这些变量就行了。
